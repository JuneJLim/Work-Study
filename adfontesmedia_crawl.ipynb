```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from webdriver_manager.chrome import ChromeDriverManager

from bs4 import BeautifulSoup

import time
import csv

import re

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url = "https://www.adfontesmedia.com/rankings-by-individual-news-source/"
driver.get(url)
```

```python
full_info = []

page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')

info_list = soup.find_all('h2', class_='blog-shortcode-post-title entry-title fusion-responsive-typography-calculated')

for item in info_list:
    info = []
    info.append(item.get_text())
    # if there is a hyperlink (to the actual ranking info page), store it
    if item.find('a') != None:
        info.append(item.find('a').get('href'))
    
    full_info.append(info)
```

```python
driver.get('https://www.adfontesmedia.com/vox-bias-and-reliability/')

page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')

info_list = soup.find_all('b')
url_list = soup.find_all('a')
```

```python
final_info_2 = []

for i in full_info[106:]:
    driver.get(i[1])
    
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')

    info_list = soup.find_all(['b', 'strong'])
    url_list = soup.find_all('a')
    
    
    url = url_list[106].get('href')
    name = info_list[0].get_text()[:-30]
    bias = info_list[1].get_text().strip()
    rel = info_list[2].get_text().strip()
    try:
        bias_score = info_list[5].get_text().strip()[6:]
    except IndexError:
        bias_score = None
    try:
        rel_score = info_list[4].get_text().strip()[13:]
    except IndexError:
        rel_score = None
    
    final_info_2.append([url, name, bias, rel, bias_score, rel_score])
    time.sleep(1)
```

```python
final_info_full = final_info + final_info_2
```

```python
with open("adfontes_data.csv", "w", newline="", encoding='utf-16') as f:
    writer = csv.writer(f, delimiter='\t')
    for s in final_info_full:
        writer.writerow(s)
```
